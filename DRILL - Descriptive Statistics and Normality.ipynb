```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2>Gamma</h2>

```python
distributionvar = np.random.gamma(5, 1, 100)

plt.hist(distributionvar)

plt.axvline(distributionvar.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(distributionvar.mean() - distributionvar.std(), color = 'b', linestyle='dashed', linewidth=2)

plt.axvline(distributionvar.mean() + distributionvar.std(), color = 'b', linestyle='dashed', linewidth=2)

plt.show()
```

<h2>Normal</h2>

```python
distributionvar = np.random.normal(size=100)

plt.hist(distributionvar)

plt.axvline(distributionvar.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(distributionvar.mean() - distributionvar.std(), color = 'b', linestyle='dashed', linewidth=2)

plt.axvline(distributionvar.mean() + distributionvar.std(), color = 'b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
norm1 = np.random.normal(5, .5, 100)

norm2 = np.random.normal(10, 1, 100)

norm1.sort()

norm2.sort()

norm3 = norm1 + norm2

mean = np.mean(norm3)

sd = np.std(norm3)

plt.hist(norm3)

plt.axvline(mean, color='b', linewidth=2)

plt.axvline(mean - sd, color='b', linewidth=2, linestyle='dashed')

plt.axvline(mean + sd, color='b', linewidth=2, linestyle='dashed')

plt.show()
```
